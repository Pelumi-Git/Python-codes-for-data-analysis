```python
#correlation analysis 
# ============================================================================
# CORRELATION ANALYSIS: STUDY HOURS PER WEEK vs FINAL GRADE
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Step 1: Calculate Correlation Coefficient
correlation = df['StudyHoursPerWeek'].corr(df['FinalGrade'])
print("=" * 60)
print("CORRELATION ANALYSIS")
print("=" * 60)
print(f"\nCorrelation Coefficient: {correlation:.4f}")

# Step 2: Perform Pearson Correlation Test
# Remove NaN values before correlation
valid_data = df[['StudyHoursPerWeek', 'FinalGrade']].dropna()
r_value, p_value = pearsonr(valid_data['StudyHoursPerWeek'], valid_data['FinalGrade'])
print(f"Pearson R-value: {r_value:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significant? {'YES (p < 0.05)' if p_value < 0.05 else 'NO (p >= 0.05)'}")

# Step 3: Interpretation
if abs(correlation) >= 0.7:
    strength = "Very Strong"
elif abs(correlation) >= 0.5:
    strength = "Strong"
elif abs(correlation) >= 0.3:
    strength = "Moderate"
else:
    strength = "Weak"

print(f"\nCorrelation Strength: {strength}")
print(f"Direction: {'Positive' if correlation > 0 else 'Negative'}")
print("=" * 60)

# Step 3: Interpretation
if abs(correlation) >= 0.7:
    strength = "Very Strong"
elif abs(correlation) >= 0.5:
    strength = "Strong"
elif abs(correlation) >= 0.3:
    strength = "Moderate"
else:
    strength = "Weak"

print(f"\nCorrelation Strength: {strength}")
print(f"Direction: {'Positive' if correlation > 0 else 'Negative'}")
print("=" * 60)

# Step 4: Group Analysis by Study Hours Categories
def categorize_study(hours):
    if hours >= 15:
        return 'Intensive (15+)'
    elif hours >= 10:
        return 'High (10-14)'
    elif hours >= 5:
        return 'Moderate (5-9)'
    else:
        return 'Low (<5)'

df['StudyCategory'] = df['StudyHoursPerWeek'].apply(categorize_study)

# Summary statistics by study category
print("\nGRADE STATISTICS BY STUDY HOURS CATEGORY:")
print("-" * 60)

grouped = df.groupby('StudyCategory')['FinalGrade'].agg([
    ('Count', 'count'),
    ('Mean Grade', 'mean'),
    ('Median', 'median'),
    ('Std Dev', 'std'),
    ('Min', 'min'),
    ('Max', 'max')
]).round(2)

print(grouped)
print("-" * 60)

# Step 5: Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Scatter plot with trend line
ax1.scatter(df['StudyHoursPerWeek'], df['FinalGrade'], alpha=0.6, s=100, color='steelblue')

# Add regression line
# Remove NaN values for visualization
valid_data = df[['StudyHoursPerWeek', 'FinalGrade']].dropna()

# Add regression line
z = np.polyfit(valid_data['StudyHoursPerWeek'], valid_data['FinalGrade'], 1)
p = np.poly1d(z)
x_line = np.linspace(valid_data['StudyHoursPerWeek'].min(), valid_data['StudyHoursPerWeek'].max(), 100)

ax1.set_xlabel('Study Hours Per Week', fontsize=12, fontweight='bold')
ax1.set_ylabel('Final Grade', fontsize=12, fontweight='bold')
ax1.set_title('Study Hours vs Final Grades', fontsize=13, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Box plot by category
categories = ['Low (<5)', 'Moderate (5-9)', 'High (10-14)', 'Intensive (15+)']
data_to_plot = [df[df['StudyCategory'] == cat]['FinalGrade'].values 
                 for cat in categories if cat in df['StudyCategory'].unique()]

ax2.boxplot(data_to_plot, tick_labels=[cat for cat in categories if cat in df['StudyCategory'].unique()])
ax2.set_ylabel('Final Grade', fontsize=12, fontweight='bold')
ax2.set_title('Grade Distribution by Study Category', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# Step 6: Summary and Conclusions
print("\n" + "=" * 60)
print("SUMMARY & CONCLUSIONS")
print("=" * 60)

print(f"\n1. CORRELATION COEFFICIENT: {correlation:.4f}")
print(f"   → {strength} {('positive' if correlation > 0 else 'negative')} correlation")
print(f"   → {'Statistically significant at p < 0.05' if p_value < 0.05 else 'Not statistically significant'}")

print(f"\n2. LINEAR RELATIONSHIP:")
print(f"   → For every 1 additional study hour:")
print(f"   → Final grade increases by {z[0]:.2f} points")

print(f"\n3. GRADE IMPROVEMENT BY CATEGORY:")
categories_ordered = ['Low (<5)', 'Moderate (5-9)', 'High (10-14)', 'Intensive (15+)']
for cat in categories_ordered:
    if cat in df['StudyCategory'].unique():
        avg = df[df['StudyCategory'] == cat]['FinalGrade'].mean()
        count = len(df[df['StudyCategory'] == cat])
        print(f"   → {cat}: {avg:.2f} avg ({count} students)")

print(f"\n4. KEY FINDING:")
if abs(correlation) < 0.3:
    print("   → Study hours show a WEAK relationship with final grades")
    print("   → Quality of study matters more than quantity")
    print("   → Other factors (attendance, effort, etc.) may be more important")
elif abs(correlation) < 0.7:
    print("   → Study hours show a MODERATE relationship with final grades")
    print("   → Increasing study time can help, but other factors matter too")
else:
    print("   → Study hours show a STRONG relationship with final grades")
    print("   → Encouraging more study hours can significantly improve performance")

print("\n" + "=" * 60)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Remove NaN values
valid_data = df[['StudyHoursPerWeek', 'FinalGrade']].dropna()

# Prepare data for model
X = valid_data['StudyHoursPerWeek'].values.reshape(-1, 1)  # Independent variable (Study Hours)
y = valid_data['FinalGrade'].values  # Dependent variable (Final Grade)

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Get model parameters
slope = model.coef_[0]
intercept = model.intercept_

# Calculate model performance
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

print("=" * 70)
print("LINEAR REGRESSION MODEL: Study Hours → Final Grade")
print("=" * 70)
print(f"\nModel Equation: Final Grade = {slope:.4f} × StudyHours + {intercept:.4f}")
print(f"\nModel Performance:")
print(f"  • R-squared (R²): {r2:.4f}  (Explains {r2*100:.2f}% of variance)")
print(f"  • RMSE: {rmse:.2f} points (Average prediction error)")
print("=" * 70)
```

```python
# Calculate average study hours
avg_study_hours = df['StudyHoursPerWeek'].mean()

# Filter students with above-average study hours
above_avg_students = df[df['StudyHoursPerWeek'] > avg_study_hours].copy()

# Make predictions for above-average students
above_avg_students['PredictedGrade'] = model.predict(above_avg_students[['StudyHoursPerWeek']])

print("\n" + "=" * 70)
print("STUDENTS WITH ABOVE-AVERAGE STUDY HOURS")
print("=" * 70)
print(f"\nAverage Study Hours (All Students): {avg_study_hours:.2f} hours/week")
print(f"Number of Above-Average Students: {len(above_avg_students)} out of {len(df)}")

# Display results table
results_display = above_avg_students[[
    'StudentID', 'Name', 'StudyHoursPerWeek', 'FinalGrade', 'PredictedGrade'
]].copy()

results_display.columns = ['ID', 'Name', 'Study Hours', 'Actual Grade', 'Predicted Grade']
results_display['Study Hours'] = results_display['Study Hours'].round(2)
results_display['Actual Grade'] = results_display['Actual Grade'].round(2)
results_display['Predicted Grade'] = results_display['Predicted Grade'].round(2)

print(results_display.to_string(index=False))
```

```python
# Calculate prediction errors
above_avg_students['PredictionError'] = above_avg_students['FinalGrade'] - above_avg_students['PredictedGrade']
above_avg_students['AbsoluteError'] = abs(above_avg_students['PredictionError'])

mae = above_avg_students['AbsoluteError'].mean()  # Mean Absolute Error
rmse_above = np.sqrt((above_avg_students['PredictionError']**2).mean())

print("\n" + "=" * 70)
print("PREDICTION ACCURACY FOR ABOVE-AVERAGE STUDENTS")
print("=" * 70)
print(f"\n  • Mean Absolute Error (MAE): {mae:.2f} points")
print(f"  • Average Prediction Error: ±{mae:.2f} points")
print(f"  • Average Actual Grade: {above_avg_students['FinalGrade'].mean():.2f}")
print(f"  • Average Predicted Grade: {above_avg_students['PredictedGrade'].mean():.2f}")
```
