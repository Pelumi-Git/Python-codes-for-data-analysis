```python
# Sales trend by month
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])
df['Month'] = df['Sale_Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales_Amount'].sum().sort_index()
monthly_sales
```

```python
#FORECAST 2 YEARS (24 MONTHS) SALES
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import numpy as np

# Prepare data for forecasting
sales_values = monthly_sales.values

# Fit Exponential Smoothing model
model = ExponentialSmoothing(sales_values, trend='add', seasonal=None)
fitted_model = model.fit()

# Forecast next 24 months (2 years)
forecast_24months = fitted_model.forecast(steps=24)

# Create forecast dataframe
forecast_df = pd.DataFrame({
    'Month': pd.period_range(start=monthly_sales.index[-1] + 1, periods=24, freq='M'),
    'Forecasted_Sales': forecast_24months
})

print("2-Year Revenue Forecast (24 Months):")
print(forecast_df)
print(f"\nTotal Forecasted Revenue for Next 2 Years: {forecast_24months.sum():,.2f}")
```

```python
# Yearly Sales with Group By
df['Year'] = df['Sale_Date'].dt.year
yearly_sales = df.groupby('Year')['Sales_Amount'].sum().sort_index()
print("Yearly Sales Summary:")
print(yearly_sales)
yearly_sales
```
